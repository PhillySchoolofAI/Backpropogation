# Introduction to Neural Networks

## Perceptrons
A Perceptron is an artificial neuron. The perceptron algorithm was invented in 1957 at the Cornell Aeronautical Laboratory by Frank Rosenblatt and was funded by the United States Office of Naval Research. It was inspired by earlier work by Warren McCulloch and Walter Pitts.

## How do Perceptrons Work?
A perceptron takes several binary inputs (x<sub>1</sub>, x<sub>2</sub>, ... x<sub>n</sub>) and produces a single binary output.


## Weights
(w<sub>1</sub>, w<sub>2</sub>, ... w<sub>n</sub>)

Real numbers expressing the importance of the respective inputs to the output.

The neuron's output, 0 or 1, is determined by whether the weighted sum,

∑<sub>j</sub>w<sub>j</sub>x<sub>j</sub>

is less than or greater than some threshold value. Just like weights the threshold is a real number which is a parameter of the neuron.



## Sigmoid Neurons
The main model of artificial neurons used in modern work on neural networks is called the sigmoid neuron.
